# Прогноз оттока клиентов

**Цель исследования:** научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. 

**Задача исследования:** построить модель машинного обучения, которая поможет прогнозировать отток клиентов. 

Данные для исследования состоят из файлов, полученных из разных источников: 

- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец customerID содержит код клиента.
Информация о договорах актуальна на 1 февраля 2020.

**Описание услуг**

1. Оператор предоставляет два основных типа услуг:

- Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

2. Также доступны такие услуги:

- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание данных**

- *BeginDate* – дата начала пользования услугами;
- *EndDate* – дата окончания пользования услугами;
- *Type* – тип оплаты: ежемесячный, годовой и тд;
- *PaperlessBilling* – электронный платёжный документ;
- *PaymentMethod* – способ оплаты;
- *MonthlyCharges* – ежемесячные траты на услуги по договору;
- *TotalCharges* – всего потрачено денег на услуги;
- *Dependents* – наличие иждивенцев;
- *Senior Citizen* – наличие пенсионного статуса по возрасту;
- *Partner* – наличие супруга(и);
- *MultipleLines* – наличие возможности ведения параллельных линий во время звонка.

**Ход исследования:**

1. Загрузим данные.
2. Проведем разведочный анализ данных (EDA).
3. По результатам EDA выполним подготовку данных к машинному обучению.
4. Сформируем подробный план поэтапный план работ для дальнейшего исследования.
5. Обучим и проверим модели (от 2х и более), выполним подбор гиперпараметров.
6. Выберем одну лучшую модель и проверим ее на тестовой выборке. Значение ROC-AUC должно быть не менее 0.85.
7. Напишем общий отчет и рекомендации.

```python
def graph_kat(data, data_label1, data_label2=''):
   
    sns.set()
    f, axes = plt.subplots(1, 2, figsize=(16, 4))
    axes[0].set_title(data_label1 + ". Количество по категориям", fontsize=18)
    g = sns.countplot(data=data, x=data_label1,  hue='Target', ax=axes[0], palette='pastel')
    g.set_xticklabels(g.get_xticklabels(), rotation=90)
    
    if data_label2 != '':
        axes[1].set_title(data_label2 + ". Количество по категориям", fontsize=18)
        g = sns.countplot(data=data, x=data_label2, hue='Target', ax=axes[1], palette='pastel')
        g.set_xticklabels(g.get_xticklabels(), rotation=90)
        
    plt.show()
```

```python
def list_services (row):
    
    services = []
    
    try:
        if row['MultipleLines'] == 'No' or row['MultipleLines'] == 'Yes':
            services.append('phone')
        if  row['InternetService'] == 'fiber optic' or row['InternetService'] == 'dsl': 
            services.append('internet') 
        if  row['OnlineSecurity'] == 'Yes':
            services.append('online_security')
        if  row['DeviceProtection'] == 'Yes':
            services.append('device_protection')
        if  row['OnlineBackup'] == 'Yes':
            services.append('online_backup') 
        if  row['TechSupport'] == 'Yes':
            services.append('support')
        if  row['StreamingTV'] == 'Yes':
            services.append('streaming_tv') 
        if row['StreamingMovies'] == 'Yes':
            services.append('movies')
            
        return ', '.join(services) 
    
    except: 
        return 'ошибка'
```

```python
df['ListServices'] = df.apply(list_services, axis=1)
```

### Построение моделей и оценка их качества

Будут подобраны параметры и построены следующие модели:

- LogisticRegression;
- RandomForestClassifier;
- CatBoostClassifier.

В завершении будет выбрана модель с наилучшими показателями целевой метрики *ROC-AUC*.

Напишем функцию для подбора гиперпараметров

```python
def best_hyper_model(model, 
                     parameters_m, 
                     transform, 
                     features=features_train, 
                     target=target_train):
    
    steps = [('transform', transform), ('model', model)]
    model = Pipeline(steps)

    # подбор лучших гиперпараметров модели с учетом метрики AUC-ROC

    rand_m = RandomizedSearchCV(estimator=model, 
                                param_distributions=parameters_m, 
                                cv=5, 
                                scoring='roc_auc', 
                                n_jobs=-2, 
                                random_state=RANDOM_STATE)
    
    rand_m.fit(features, target)

    best_params = rand_m.best_params_
    print('Гиперпараметры лучшей модели', best_params)
    
    best_model = rand_m.best_estimator_
    
    best_score = rand_m.best_score_
    print('ROC-AUC модели', best_score)

    return  best_model, best_params, best_score
```

```python
prediction_test = model_CBC.predict(features_test)

probabilities_test = model_CBC.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]

roc_auc_test = roc_auc_score(target_test, probabilities_one_test)
accuracy_test = accuracy_score(target_test, prediction_test)
recall_test = recall_score(target_test, prediction_test)

print("Результат тестовой метрики ROC-AUC для модели CatBoostClassifier:", round(roc_auc_test, 4))
print("Результат тестовой метрики Accuracy для модели CatBoostClassifier:", round(accuracy_test, 4))
print("Результат тестовой метрики Recall для модели CatBoostClassifier:", round(recall_test, 4))
```

На тестовой выборке получили следующие значения:

- ROC-AUC 0.88
- Accuracy 0.81
- Recall 0.80

**ROC-AUC** можно интерпретировать как показатель модели чьи прогнозы предсказания верны на 88%. Причем AUC не зависит от масштаба - он измеряет, насколько хорошо ранжируются прогнозы, а не их абсолютные значения и также не зависит от порога классификации - он измеряет качество прогнозов модели независимо от выбранного порога классификации.

**Accuracy** — это показатель, который описывает общую точность предсказания модели по всем классам. Это особенно полезно, когда каждый класс одинаково важен. Он рассчитывается как отношение количества правильных прогнозов к их общему количеству. Стоит учесть, что метрика **Accuracy** может быть обманчивой. Один из таких случаев — это несбалансированные данные. Поскольку при дисбалансе большинство семплов принадлежит к одному классу, **Accuracy** для этого класса будет выше, чем для другого.

Полнота **Recall** выявляет, какую долю положительных среди всех ответов выделила модель. Полнота — это доля TP-ответов среди всех, у которых истинная метка 1. Хорошо, когда значение **Recall** близко к единице: модель хорошо ищет положительные объекты. 

### Построим ROC кривую.

```python
def make_roc_curve(target_test, predicted_test):
    fpr, tpr, thresholds = roc_curve(target_test, predicted_test)

    plt.figure(figsize=(10, 10))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([-0.1, 1.0])
    plt.ylim([0.0, 1.1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
    
make_roc_curve(target_test, probabilities_one_test)
```

- `True Negative` (истинно-отрицательное решение): клиент ушел из компании и прогноз модели это предсказал *(Предсказания 1 : Целевые значения 1)* - 80%

**Accuracy** — это показатель, который описывает общую точность предсказания модели по всем классам. Стоит учесть, что метрика может быть обманчивой при несбалансированности данных. Поскольку при дисбалансе большинство значений принадлежит к одному классу, метрика для этого класса будет выше, чем для другого.

Полнота **Recall** выявляет, какую долю положительных среди всех положительных ответов выделила модель. Полнота — это доля TP-ответов среди всех, у которых истинная метка 1. Хорошо, когда значение **Recall** близко к единице: модель хорошо ищет положительные объекты (в нашем случае положительный ответ имеет метку 1 - вероятность того, что клиент уйдет).

**2. Построили матрицу ошибок:**



По Матрице ошибок видна следующая картина:

- `True Positive` (истинно-положительное решение): клиент не ушел из компании и прогноз модели это предсказал *(Предсказания 0 : Целевые значения 0)* - 81%
- `False Positive` (ложноположительное решение): ошибка 1-го рода, ML-модель предсказала что клиент уйдет из компании, а на самом деле он не ушел *(Предсказания 1 : Целевые значения 0)* - 19%
- `False Negative` (ложноотрицательное решение): ошибка 2-го рода – ML-модель предсказала, что клиент останется в компании, но на самом деле он ушел *(Предсказания 0 : Целевые значения 1)* - 20%
- `True Negative` (истинно-отрицательное решение): клиент ушел из компании и прогноз модели это предсказал *(Предсказания 1 : Целевые значения 1)* - 80%

Нам важно минимизировать ложноотрицательные ошибки (0.20) (клиенты были классифицированы как лояльные, но фактически уйдут из компании), так как по данным маркетологов удержание текущих клиентов дешевле, чем возврат ушедших клиентов. Поэтому важно повысить полноту определения целевого класса 1.

**3. Оценка важности признаков:**



Видим, что наиболее важным признаком того, что клиент собирается уходить является значение длительности контракта `UsingDays` - 55.7%. Менее значимые `ListServices, PaymentMethod, MultipleLine и Partner` - 15.7%, 10%, 9.6% и 4.7% соответственно.

- Срок сотрудничества `UsingDays` - самым весомым признаком является срок сотрудничества - чем больше срок сотрудничества, тем ниже вероятность ухода клиента.

- Список используемых сервисов `ListServices` - склонны к уходу клиенты с минимальным количеством услуг - ранее мы выяснили, что это клиенты использующие только телефонную связь. Тем не менее при максимальном количестве услуг вероятность того, что клиент уйдет ниже.

- Способ оплаты `PaymentMethod` - пользователи, которые используют почтовый способ оплаты менее склонны к уходу.

- Наличие возможности ведения параллельных линий во время звонка `MultipleLine` - клиенты с данной услугой склонны к уходу.

- Наличие супруга/супруги `Partner` - такие клиенты также склонны к уходу.
