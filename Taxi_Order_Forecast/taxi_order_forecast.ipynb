**Сезонность (англ. seasonality)** — циклично повторяющиеся закономерности во временном ряду.

**Остаток (англ. residuals)** - это компонента, которая не объясняется трендом и сезонностью, это шум.

### Разности временного ряда.

**Разности временного ряда (англ. time series difference)** — это набор разностей между соседними элементами временного ряда, т.е. из каждого значения вычитается предыдущее.

### Создание признаков.

Создадим признаки для горизонта прогнозирования в один шаг.

Напишем функцию `make_features` для создания признаков. Разберём каждый тип признаков:

1. Календарные признаки (англ. calendar features) - судя по сезонности, нам не так важны месяц и год, как день недели, день и час заказа.
2. «Отстающие значения» (англ. lag features) - предыдущие значения временного ряда подскажут, будет ли функция расти или уменьшаться. Получим отстающие значения знакомой функцией *shift()*.
3. Скользящее среднее - как признак задаёт общий тренд временного ряда. Скользящее среднее *rolling_mean* с шириной окна *rolling_mean_size*.

```python
def make_features(data, max_lag, rolling_mean_size):
    
    data_new = data.copy()
    
    data_new['day'] = data.index.day
    data_new['dayofweek'] = data.index.dayofweek 
    data_new['hour'] = data.index.hour # календарные признаки
    
    for lag in range(1, max_lag + 1):
        data_new['lag_{}'.format(lag)] = data_new['num_orders'].shift(lag) # отстающие значения

    data_new['rolling_mean'] = data_new['num_orders'].shift().rolling(rolling_mean_size).mean() # скользящее среднее
    return data_new
```

Так как мы имеем суточную и недельную сезонность, размер окна скользящего среднего(rolling_mean_size) и число отстающих значений(max_lag ) возьмем равными 168(столько часов в неделе).

```python
warnings.filterwarnings('ignore')
data_new = make_features(data, 168, 168)
data_new.head(15)
```

Получили целый набор признаков, которые теперь можно использовать для прогнозирования.

## Обучение моделей

Выберем следующие алгоритмы для обучения модели:

- `LinearRegression`
- `RandomForestRegressor`
- `CatBoostRegressor`
- `LGBMRegressor`

Создадим метрику, с помощью которой будет определяться работа модели - *RMSE_scorer*.

```python
def RMSE_scorer(model, features_train, target_train, features_valid, target_valid):
    
    model.fit(features_train, target_train)
     
    pred_valid = model.predict(features_valid)
    
    RMSE_valid = mean_squared_error(target_valid, pred_valid, squared=False)

    print('RMSE модели на валидационной выборке:', RMSE_valid)
    
    return RMSE_valid
```

Подбор лучших гиперпараметров осуществим с помощью RandomizedSearchCV. Воспользуемся кросс-валидацией для временных рядов *TimeSeriesSplit()*

```python
tscv = TimeSeriesSplit(n_splits=3)
```

```python
def best_hyper_model(model, parameters_m, features, target):
    
    # подбор лучших гиперпараметров модели с учетом метрики RMSE
    
    grid_m = RandomizedSearchCV(model, parameters_m, cv=tscv, n_jobs=-1, scoring='neg_root_mean_squared_error', random_state=12345)
    grid_m.fit(features, target)

    best_params = grid_m.best_params_

    print('Гиперпараметры лучшей модели', best_params)
    
    # обучение модели с лучшими гиперпараметрами
    # функция возвращает модель, гиперпараметры
    
    best_model = model
    best_model.set_params(**best_params) # добавление лучших гиперпараметров

    return best_model, best_params
```

### Линейная регрессия (LinearRegression)

```python
parameters_LR ={'fit_intercept':[True, False], 
             'copy_X':[True, False]}
```

```python
%%time

warnings.filterwarnings('ignore')
model_LR, params_LR = best_hyper_model(
    LinearRegression(), 
    parameters_LR, 
    features_train,
    target_train
)
```

```python
RMSE_scorer_LR = RMSE_scorer(
    model_LR, features_train, target_train, features_valid, target_valid
)
```
